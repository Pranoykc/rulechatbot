```python
import re
import random

class Rulebot:
    negative_responses = ('no', 'nope', 'nah', 'not a chance', 'sorry', 'never')
    exit_commands = ('quit', 'pause', 'exit', 'goodbye', 'bye', 'later')
    help_commands = ('help', 'instructions', 'what can you do', 'commands')
    random_ques = (
        'Why are you here?\n',
        'Are there any humans like you?\n',
        'Does Earth have a leader?\n',
        'What planets have you visited?\n'
    )

    def __init__(self):
        self.alienbabble = {
            'describe_planet_earth': r'.*\s*your planet.*',
            'answer_why_intent': r'why\s*are.*',
            'about_intellipat': r'.*\s*intellipaat',
            'answer_from_intent': r'from\s*where.*',
            'favorite_food': r'.*\s*your favorite food.*',
            'hobbies': r'.*\s*your hobbies.*'
        }
        self.user_details = {}
        self.conversation_log = []

    def greet(self):
        self.user_details['name'] = input('What is your name?\n')
        self.conversation_log.append(f"User: {self.user_details['name']}")
        will_help = input(f"Hi {self.user_details['name']}, I am Alicee. Will you help me learn about your planet?\n")
        self.conversation_log.append(f"User: {will_help}")
        if will_help.lower() in self.negative_responses:
            print('Ok, have a nice Earth day!')
            self.conversation_log.append("Bot: Ok, have a nice Earth day!")
            return
        self.chat()

    def make_exit(self, reply):
        if reply in self.exit_commands:
            print('Ok, have a nice Earth day!')
            self.conversation_log.append("Bot: Ok, have a nice Earth day!")
            return True
        return False

    def chat(self):
        reply = input(random.choice(self.random_ques)).lower()
        self.conversation_log.append(f"Bot: {reply}")
        while not self.make_exit(reply):
            reply = input(self.match_reply(reply)).lower()
            self.conversation_log.append(f"User: {reply}")

    def match_reply(self, reply):
        for key, value in self.alienbabble.items():
            intent = key
            regex_pattern = value
            found_match = re.match(regex_pattern, reply)
            if found_match:
                response_function = getattr(self, intent)
                return response_function()
        if reply in self.help_commands:
            return self.provide_help()
        return self.no_match_intent()

    def describe_planet_earth(self):
        responses = ('My planet is Mars.\n', 'I am the guardian of galaxy Neptivia.\n')
        response = random.choice(responses)
        self.conversation_log.append(f"Bot: {response}")
        return response

    def answer_why_intent(self):
        responses = ('I came here to destroy the Earth.\n', 'I came here to rule Earth.\n', 'I came here to see the nature.\n')
        response = random.choice(responses)
        self.conversation_log.append(f"Bot: {response}")
        return response

    def about_intellipat(self):
        responses = ('Intellipaat is a great institute.\n', 'Intellipaat will make your career grow.\n', 'Intellipaat is a professional course platform.\n')
        response = random.choice(responses)
        self.conversation_log.append(f"Bot: {response}")
        return response

    def answer_from_intent(self):
        responses = ('I am from Pluto.\n', 'I came from 200,000 km away.\n')
        response = random.choice(responses)
        self.conversation_log.append(f"Bot: {response}")
        return response

    def favorite_food(self):
        responses = ('I like cosmic dust.\n', 'Space berries are my favorite.\n')
        response = random.choice(responses)
        self.conversation_log.append(f"Bot: {response}")
        return response

    def hobbies(self):
        responses = ('I like to observe stars.\n', 'Exploring galaxies is my hobby.\n')
        response = random.choice(responses)
        self.conversation_log.append(f"Bot: {response}")
        return response

    def no_match_intent(self):
        responses = ('Tell me more.\n', 'Why do you say that?\n', 'Can you elaborate?\n', 'Why?\n')
        response = random.choice(responses)
        self.conversation_log.append(f"Bot: {response}")
        return response

    def provide_help(self):
        response = "I can answer questions about why I'm here, where I'm from, and tell you about Intellipaat. Just ask!"
        self.conversation_log.append(f"Bot: {response}")
        return response

Alienbot = Rulebot()
Alienbot.greet()


```
